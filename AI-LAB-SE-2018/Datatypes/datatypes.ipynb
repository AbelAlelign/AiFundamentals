## Determining the type of data using the type() method.

- In Python, the type() function is used to determine the data type of a value or variable.
Syntax
- type(object)
- object can be a variable, value, or expression.

- The function returns the data type.

        Examples
        - a = 10
        - print(type(a))        # <class 'int'>

        - b = 3.14
        - print(type(b))        # <class 'float'>

        - c = "Hello"
        - print(type(c))        # <class 'str'>

        - num = 2.5j
        - print(type(num))      # <class 'complex'>


```python
a= 10
print(type(a))
name="getabalew"
print(type(name))
```

```python
b= 2.5j
print(type(b))
```

## Compound data types (list, tuple, set and dictionary)

- Compound data types in Python are data types that can hold multiple values in a single variable. 
- Unlike simple data types (like int, float, or bool), compound types can store collections of data.

### List
- Ordered, Mutable Collection
- A list is an ordered collection of items that can hold different data types. 
- Lists are mutable (modifiable).
     - Elements have a fixed order.
     - Can be modified (add, remove, change elements).
     - Allows Duplicates – Can contain repeated elements.

- square braces [ ]

```python
# Example
list1 = ["Keyboard", "RAM", "RAM", 5, "Mouse"]
list1[0] = "Printer"
print(list1)
```

```python
# creating list
dept = ["IT", "IS", "DS", "CS"]
print(dept)

    # Accessing elements index 0
print(f" Data 1: {dept[0]}")

    # Modifying data index 1 with "information systems"
dept[1]= "information systems"
print(f" list after modifcation: {dept}")

    # add list(SW) item at the end of the list 
dept.append("sw")
print(f" List after appened: {dept}")

# insert data "Software" at any position
dept.insert(2,"software")
print(f" List after insert :{dept}")

# remove data "Software"  from the list 
dept.remove("software")
print(f" List after remove: {dept}")

# extend list or merge two lists [4,5,6,7]
dept.extend([4,5,6,7])
print(f" list after extend: {dept}")
```

### Tuple
- A tuple is an ordered, immutable collection in Python that allows storing multiple items in a single variable.
- Unlike lists, tuples cannot be modified after creation, making them a great choice for storing constant data.
- It can contain repeated data items
- It supports multiple data types on the same collection
- round braces ( )

```python
# Example
tuple1 = ("Keyboard", "RAM","RAM", "CPU", "Mouse")
# tuple1[0] = "Printer"
print(tuple1)
```

### Set
- A set in Python is an unordered collection of unique elements.
- Sets are commonly used when you need to store multiple items and avoid duplicates.
- Unlike lists or tuples, sets do not allow indexing, slicing, or other sequence-like behavior.
- Its Possible to create a set by placing values inside curly braces  {}  or by using  set()  method.
- we can use set:
     - when we want to remove duplicates from a collection and don't care about the order
     - when we want to perform set operations like union, intersection, difference


```python
# Example
set1 = set(["Keyboard", "RAM","RAM", "CPU", "Mouse"])
#tuple1[0] = "Printer"
# set[0] = "Printer"
print(set1)
```

```python
set1={"Keyboard", "RAM","RAM", "CPU", "Mouse"}
print(set1)
```

### Dictionary
- Python dictionaries are a powerful built-in data structure that allows you to store key-value pairs for efficient data retrieval and manipulation.
- Dictionaries are defined using curly braces { } and consist of keys and corresponding values. 
- Basic syntax:
     dict = {'key1':'value1', 'key2': 'value2', 'key3':'value3'}

```python
capital_city ={
    'Ethiopia':'Addis Ababa',
    'Kenya':'Nairobi',
    'USA':'WDC'  
}
print(capital_city)
print(capital_city["Kenya"])
```
