# 2. Python Comments

- Comments are notes or explanations in your code that are ignored by the Python interpreter.

- They are used to make the code more readable for humans.

- Python does not execute comments—they are only for documentation.

1. Starts with #. Every thing after # on that line is ignored.
2. triple quotes (''' ... ''' or """ ... """) for multi-line comments or doc strings.

```python
# This is a single-line comment
print("Hello, World!")  # This prints a message
```

# 4. Python String Methods

- Python has a set of built-in methods that you can use on strings.
- Python string is a sequence of Unicode characters that is enclosed in quotation marks

```python
# Converts all uppercase characters in a string into lowercase.

text = "HELLO World"
print(text.lower())  
```

```python
#Converts all lowercase characters in a string into uppercase.

text = "Hello World"
print(text.upper())  
```

```python
# Converts the first character of each word to uppercase (title case).

text = "python programming"
print(text.title())  
```

```python
# Swaps the case of all characters in the string.

text = "Python Is Fun"
print(text.swapcase()) 
```

```python
# Converts only the first character of the string to uppercase.

text = "python "
print(text.capitalize())  
```

```python
# Returns True if all characters in the string are digits.

text = "12345"
print(text.isdigit())  
```

```python
# Returns True if all characters in the string are uppercase.

text = "HELLO"
print(text.isupper())  
```

```python
# Returns True if all characters in the string are lowercase.

text = "hello"
print(text.islower())  
```

```python
# Returns the length of the string (number of characters).

text = "Python "
print(len(text))  
```

# 7. Conditional and loop statments in Python

- Loop statements execute a sequence of statements multiple times.
- A conditional statement tells a program to execute an action depending on whether a condition is true or false.

# if else statments

```python
# Example: python code to check if a Number is Positive, Negative or 0 using if else statment

num = int(input("Enter any number:"))
if num>0:
    print("The number is +ve")
elif num==0:
    print("The number is zero")
else:
    print("The number is -ve")
```

# for loop statments

```python
# Example: Write a Python program to print numbers from 1 to 10 using for loop. 

for num in range(3,7):
    print(num)
```

## continue and break 
- continue and break are control statements used inside loops, but they work differently:
- continue: Skips the current iteration and moves to the next iteration of the loop
- break: Completely exits the loop immediately

```python
#Example of continue
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

```python
# Example of break
for i in range(1, 6):
    if i == 3:
        break
    print(i)
```

# While loop statments

- A while loop is used to repeat a block of code as long as a given condition is true.



```python
# Python Program to Print Numbers From 1 to 10 Using While Loop

i = 1
while(i<=10):
    print(i)
    i += 1
```
