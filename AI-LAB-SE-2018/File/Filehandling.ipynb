<h1 style="color:blue"> File Handling in Python </h1>

- Python supports file handling and allows users to handle files i.e., to read and write files, along with many other file handling options, to operate on files (file format: .txt, .jpg, csv ...)

- File Opening Modes in Python

- The general syntax for opening a file in Python is:

- f = open(filename, mode)

- r:  Opens an existing file for reading only. File must exist, otherwise error.
- w:  Opens a file for writing only. If file exists, its content is overwritten. If file does not exist, it creates a new file.
- a: Opens a file for adding data at the end. It does not delete existing content. Creates the file if it does not exist.
- r+:  Opens a file for both reading and writing. File must exist. Does not overwrite existing data automatically.

# Working with text files

## - Prepare your text file 

"r" mode

```python
file = open("file.txt", "r")
disp_content = file.read()
print(disp_content)
file.close()
```

# "w" mode

```python
file = open("file.txt", "w")
file.write("Any text")
file.close()
```

# "a" mode


```python
file = open("file.txt", "a")
file.write(" new text")
file.close()
```

# r+ mode

```python
file = open("file.txt", "r+")
disp_content = file.read()
file.write("It is new text .....")
file.close()
```

```python
file = open("file.txt", "r")
disp_content = file.read()
print(disp_content)
```

# Working with csv files

```python
csv_file = open("file.csv", "w")
csv_file.write("Name, Age, Gender\n")
csv_file.write("John, 25, Male\n")
csv_file.write("Jane, 30, Female\n")
csv_file.close()
print("CSV file created successfully.")
# read then
csv_file_reader=open("file.csv", "r")
content = csv_file_reader.read()
print(content)
csv_file_reader.close()
```
